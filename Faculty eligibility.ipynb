```python
import pandas as pd
```

```python
df = pd.read_csv("Faculty_profile.csv")
df
```

```python
df['Age'] = df['Age & DOB'].str.extract(r'(\d+)', expand=False).astype(int)
```

```python
print(df.columns)
```

```python
qual_col = 'Percentage '
```

```python
df['Qualification_Percentage'] = df[qual_col].str.extract(r'(\d+)%').astype(float)
```

```python
MIN_QUALIFICATION_PERCENTAGE = 70
MIN_EXPERIENCE_YEARS = 1
MAX_AGE = 45
```

```python
def check_eligibility(row):
    if (row['Qualification_Percentage'] >= MIN_QUALIFICATION_PERCENTAGE) and \
       (row['Experience'] >= MIN_EXPERIENCE_YEARS) and \
       (row['Age'] <= MAX_AGE):
        return "Eligible"
    else:
        return "Not Eligible"
```

```python
df['Eligibility'] = df.apply(check_eligibility, axis=1)
```

```python
df_result = df[['Name of the faculty', 'Eligibility']]
```

```python
result=pd.DataFrame(df_result)
result
```
